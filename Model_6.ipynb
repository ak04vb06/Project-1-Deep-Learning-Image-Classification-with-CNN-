```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
from PIL import Image

# v2 ARCHITECTURE & TRAINING NOTES
# ---------------------
# 1. More Depth: Increased filters (32 -> 256) to capture better details/textures.
# 2. Stability: Added BatchNormalization after Conv layers to fix unstable gradients.
# 3. Efficiency Fix: Swapped 'Flatten' for 'GlobalAveragePooling2D'.
#    - 'Flatten' created ~26M params (too heavy/overfitting).
#    - 'GlobalAvg' keeps it under 1M (~400k) and efficient.
# 4. Adaptive Learning: Added 'ReduceLROnPlateau' to automatically lower the 
#    learning rate when the model hits a plateau, allowing finer convergence.
# 5. Overfitting Control: Implemented 'EarlyStopping' to halt training and 
#    restore the best weights if validation loss stops improving.
# 6. Optimized Compilation: Configured Adam optimizer paired with dynamic 
#    callbacks for a more robust and automated training process.
# 7. Training Duration: Adjusted EPOCHS to 15 to balance training time and 
#    model convergence, relying on EarlyStopping for optimal termination.

# -----------------------------
#  CONFIGURATION (change model = change paramethers)
# -----------------------------
DATASET_PATH = r"/Users/ia_dev/Desktop/Work/archive/raw-img" # need to change 
IMG_SIZE = 224
BATCH_SIZE = 128
EPOCHS = 15
RANDOM_SEED = 42

# -----------------------------
# 1 DATA PREPARATION & METADATA
# -----------------------------
data = []

# Converts folder structure into a clean list.
for folder in os.listdir(DATASET_PATH):
    folder_path = os.path.join(DATASET_PATH, folder)
    
    if os.path.isdir(folder_path):
        # Translate label 4 the folders
        label = folder
        
        for file in os.listdir(folder_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                img_path = os.path.join(folder_path, file)
                data.append([img_path, label])

# Converts the raw list into a structured table (DataFrame)
df = pd.DataFrame(data, columns=["image_path", "label"])

# Identifies all unique categories and sorts them alphabetically
class_names = sorted(df['label'].unique())
# Creates a dictionary to map text labels to integer IDs (Model readable)
label_map = {name: i for i, name in enumerate(class_names)}
# Creates a reverse dictionary to map IDs back to text labels (Human readable)
inv_label_map = {i: name for name, i in label_map.items()}

# Map labels to integers immediately in the DataFrame
df['label_code'] = df['label'].map(label_map)

print(f"Total images: {len(df)}")
print(f"Classes found: {len(class_names)}")
print(df['label'].value_counts().head())

# -----------------------------
# 2 TRAIN / TEST SPLIT
# -----------------------------
train_df, test_df = train_test_split(
    df,
    test_size=0.1,
    stratify=df['label'], # Stratify ensures all classes are represented
    random_state=RANDOM_SEED
)

print(f"Train size: {len(train_df)} | Test size: {len(test_df)}")

# -----------------------------
# 3 DATASET PIPELINE (tf.data)
# -----------------------------
def df_to_dataset(dataframe, shuffle=True, batch_size=32):
    # Get the list of files and labels
    paths = dataframe['image_path'].values
    labels = dataframe['label_code'].values
    
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))

    # Define Image Processing (Load -> Resize -> Normalize)
    def process_path(path, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
        img = img / 255.0  # Normalize [0, 1]
        return img, label
    
    # Apply processing to all images in parallel
    ds = ds.map(process_path, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.cache()

    # Aleatory combination
    if shuffle:
        ds = ds.shuffle(buffer_size=len(dataframe))

    # Group into packs of 32
    ds = ds.batch(batch_size)
    #Preload next pack in background (No Lag)
    ds = ds.prefetch(buffer_size=tf.data.AUTOTUNE)
    return ds

# Create the final pipelines
train_dataset = df_to_dataset(train_df, batch_size=BATCH_SIZE)
test_dataset = df_to_dataset(test_df, shuffle=False, batch_size=BATCH_SIZE)

# -----------------------------
#3.5 DATA AUGMENTATION 
# -----------------------------

# We define a temporary layer just to see the effects.
aug_layer = tf.keras.Sequential([
  layers.RandomFlip("horizontal_and_vertical"),
  layers.RandomRotation(0.2),
  layers.RandomZoom(0.2),
])

plt.figure(figsize=(10, 10))
# We took 1 batch from the REAL training dataset.
for images, _ in train_dataset.take(1):
    first_image = images[0]
    
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        # We apply the transformation 9 times to the same image.
        augmented_image = aug_layer(tf.expand_dims(first_image, 0))
        
        # We display without dividing by 255 because they are already floats 0-1.
        plt.imshow(augmented_image[0])
        plt.axis('off')
        
plt.suptitle("Augmentation Preview", fontsize=16)
plt.show()

# -----------------------------
# 4 MODEL ARCHITECTURE (CNN)
# -----------------------------
# Data augmentation
data_augmentation = models.Sequential([
  layers.RandomFlip("horizontal_and_vertical"),
  layers.RandomRotation(0.2),
  layers.RandomZoom(0.2),
])

model = models.Sequential([
    # Data augmentation
    layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    data_augmentation,

    # Block 1
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),

    # Block 2
    layers.Conv2D(128, 3, padding='same', activation='relu'),
    layers.BatchNormalization(),                             #
    layers.MaxPooling2D(),

    # Block 3
    layers.Conv2D(128, 3, padding='same', activation='relu'),
    layers.BatchNormalization(),    
    layers.MaxPooling2D(),
    
    #Another block
    layers.Conv2D(256, 3, padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),

    layers.GlobalAveragePooling2D(), 

    # Dense Layers
    layers.Dense(256, activation='relu'), 
    layers.BatchNormalization(),
    layers.Dropout(0.5), 

    # Output Layer 
    layers.Dense(len(class_names), activation='softmax')
])

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', 
    factor=0.2,   
    patience=3,   
    min_lr=1e-6,  
    verbose=1
)

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=7,         
    restore_best_weights=True  
)

callbacks = [reduce_lr, early_stopping]

# 2. Compilamos el modelo
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# -----------------------------
# 5 TRAINING
# -----------------------------
history = model.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

# -----------------------------
# 6 EVALUATION & VISUALIZATION
# -----------------------------
# Evaluate on test set
test_loss, test_acc = model.evaluate(test_dataset)

train_acc = history.history['accuracy'][-1]
train_loss = history.history['loss'][-1]

print(f'Accuracy (Train): {train_acc}')
print(f'Loss (Train): {train_loss}')
print(f'Validation accuracy: {test_acc}')
print(f'Loss validation: {test_loss}')

# Plot Training History 
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss', color='blue')
plt.plot(history.history['val_loss'], label='Val Loss', color='orange') # Added validation loss
plt.title('Cross Entropy Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy', color='green')
plt.plot(history.history['val_accuracy'], label='Val Accuracy', color='red') # Added validation acc
plt.title('Classification Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# Visualize Predictions
print("\n--- Sample Predictions ---")
for images, labels in test_dataset.take(1):
    preds = model.predict(images)
    pred_classes = np.argmax(preds, axis=1)

    # Show first image in batch
    img = images[0].numpy()
    true_label = inv_label_map[labels[0].numpy()]
    pred_label = inv_label_map[pred_classes[0]]
    
    print(f"Predicted: {pred_label} | Actual: {true_label}")
    
    plt.imshow(img)
    plt.title(f"Pred: {pred_label} | True: {true_label}")
    plt.axis('off')
    plt.show()
    break
```
